```python
import numpy as np
import pandas as pd
file1 = pd.read_parquet('rep_track.parquet')
file1 = file1.set_index(np.arange(len(file1)))
```

```python
#SequenceMatcher
import difflib
def sequence_matcher(a,b):
    a_bytes = bytes(a, 'utf-8')
    b_bytes = bytes(b, 'utf-8')
    a_bytes_list = list(a_bytes)
    b_bytes_list = list(b_bytes)
    sm = difflib.SequenceMatcher(None, a_bytes_list, b_bytes_list)
    similar = sm.ratio()
    return similar
```
